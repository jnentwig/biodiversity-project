```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
observ_data = pd.read_csv("total_observations.csv")
species_data = pd.read_csv("species_info.csv")
#print(observ_data.head())
#print(observ_data.columns) #scientific_name, park_name, observations
#print(species_data.columns) #category, scientific_name, common_names, conservation_status
#What is the distribution of conservation_status for animals?
#Are the differences between species and their conservation status significant?
#Which species were spotted the most at each park?
#Are certain types of species more likely to be endangered?
#for status in species_data['conservation_status'].unique():
#    print(status)
#    print(species_data[species_data['conservation_status'] == status]['category'].value_counts())
mountain_top_species = observ_data[observ_data.park == 'Great Smoky Mountains National Park'].sort_values(by='observations', ascending=False).iloc[0:10]
yosemite_observations = []
bryce_observations = []
yellowstone_observations = []
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

#information
print(species_data.columns)

for species in mountain_top_species.species:
    yosemite_observations.append(observ_data[(observ_data.park == 'Yosemite National Park') & (observ_data.species == species)]['observations'].values[0])
    bryce_observations.append(observ_data[(observ_data.park == 'Bryce National Park') & (observ_data.species == species)]['observations'].values[0])
    yellowstone_observations.append(observ_data[(observ_data.park == 'Yellowstone National Park') & (observ_data.species == species)]['observations'].values[0])
    #get info about species
    #print(species_data[species_data.scientific_name == species][['category', 'scientific_name', 'common_names']])
    print(species_data[species_data.scientific_name == species].iloc[0, 2] + ' (' + species_data[species_data.scientific_name == species].iloc[0, 1] + ')')
    print(species_data[species_data.scientific_name == species].iloc[0, 0])
#make a graph
r = np.arange(len(mountain_top_species.species))
w = 0.2
plt.bar(r, mountain_top_species.observations, color = 'b',  width = w,  label='Great Smokey Mountains')
plt.bar(r+w, yosemite_observations, color='g', width = w, label='Yosemite National Park')
plt.bar(r+2*w, bryce_observations, color='r', width = w, label='Bryce National Park')
plt.bar(r+3*w, yellowstone_observations, color = 'y', width=w, label='Yellowstone National Park')
plt.xlabel('Species')
plt.ylabel('Number of observations')
plt.xticks(r, mountain_top_species.species, rotation=90)
plt.legend(loc='upper left')
plt.show()

#let's do the same thing for yosemite national park
yosemite_top_species = observ_data[observ_data.park == 'Yosemite National Park'].sort_values(by='observations', ascending=False).iloc[0:10]
mountain_observations = []
bryce_observations = []
yellowstone_observations = []

for species in yosemite_top_species.species:
    mountain_observations.append(observ_data[(observ_data.park == 'Great Smoky Mountains National Park') & (observ_data.species == species)]['observations'].values[0])
    bryce_observations.append(observ_data[(observ_data.park == 'Bryce National Park') & (observ_data.species == species)]['observations'].values[0])
    yellowstone_observations.append(observ_data[(observ_data.park == 'Yellowstone National Park') & (observ_data.species == species)]['observations'].values[0])
    print(species_data[species_data.scientific_name == species].iloc[0, 2] + ' (' + species_data[species_data.scientific_name == species].iloc[0, 1] + ')')
    print(species_data[species_data.scientific_name == species].iloc[0, 0])
#make a graph
r = np.arange(len(mountain_top_species.species))
w = 0.2
plt.bar(r, yosemite_top_species.observations, color = 'g',  width = w,  label='Yosemite National Park')
plt.bar(r+w, mountain_observations, color='b', width = w, label='Great Smokey Mountains')
plt.bar(r+2*w, bryce_observations, color='r', width = w, label='Bryce National Park')
plt.bar(r+3*w, yellowstone_observations, color = 'y', width=w, label='Yellowstone National Park')
plt.xlabel('Species')
plt.ylabel('Number of observations')
plt.xticks(r, yosemite_top_species.species, rotation=90)
plt.legend(loc='upper left')
plt.show()



#Bryce National Park
bryce_top_species = observ_data[observ_data.park == 'Bryce National Park'].sort_values(by='observations', ascending=False).iloc[0:10]
mountain_observations = []
yosemite_observations = []
yellowstone_observations = []

for species in yosemite_top_species.species:
    mountain_observations.append(observ_data[(observ_data.park == 'Great Smoky Mountains National Park') & (observ_data.species == species)]['observations'].values[0])
    yosemite_observations.append(observ_data[(observ_data.park == 'Yosemite National Park') & (observ_data.species == species)]['observations'].values[0])
    yellowstone_observations.append(observ_data[(observ_data.park == 'Yellowstone National Park') & (observ_data.species == species)]['observations'].values[0])
    print(species_data[species_data.scientific_name == species].iloc[0, 2] + ' (' + species_data[species_data.scientific_name == species].iloc[0, 1] + ')')
    print(species_data[species_data.scientific_name == species].iloc[0, 0])
#make a graph
r = np.arange(len(mountain_top_species.species))
w = 0.2
plt.bar(r, bryce_top_species.observations, color = 'r',  width = w,  label='Bryce National Park')
plt.bar(r+w, mountain_observations, color='b', width = w, label='Great Smokey Mountains')
plt.bar(r+2*w, yosemite_observations, color='g', width = w, label='Yosemite National Park')
plt.bar(r+3*w, yellowstone_observations, color = 'y', width=w, label='Yellowstone National Park')
plt.xlabel('Species')
plt.ylabel('Number of observations')
plt.xticks(r, bryce_top_species.species, rotation=90)
plt.legend(loc='upper left')
plt.show()



#yellowstone
yellowstone_top_species = observ_data[observ_data.park == 'Yellowstone National Park'].sort_values(by='observations', ascending=False).iloc[0:10]
mountain_observations = []
yosemite_observations = []
bryce_observations = []

for species in yosemite_top_species.species:
    mountain_observations.append(observ_data[(observ_data.park == 'Great Smoky Mountains National Park') & (observ_data.species == species)]['observations'].values[0])
    yosemite_observations.append(observ_data[(observ_data.park == 'Yosemite National Park') & (observ_data.species == species)]['observations'].values[0])
    bryce_observations.append(observ_data[(observ_data.park == 'Bryce National Park') & (observ_data.species == species)]['observations'].values[0])
    print(species_data[species_data.scientific_name == species].iloc[0, 2] + ' (' + species_data[species_data.scientific_name == species].iloc[0, 1] + ')')
    print(species_data[species_data.scientific_name == species].iloc[0, 0])
#make a graph
r = np.arange(len(mountain_top_species.species))
w = 0.2
plt.bar(r, bryce_top_species.observations, color = 'y',  width = w,  label='Yellowstone National Park')
plt.bar(r+w, mountain_observations, color='b', width = w, label='Great Smokey Mountains')
plt.bar(r+2*w, yosemite_observations, color='g', width = w, label='Yosemite National Park')
plt.bar(r+3*w, yellowstone_observations, color = 'r', width=w, label='Bryce National Park')
plt.xlabel('Species')
plt.ylabel('Number of observations')
plt.xticks(r, bryce_top_species.species, rotation=90)
plt.legend(loc='upper left')
plt.show()



#'Great Smoky Mountains National Park' 'Yosemite National Park'
# 'Bryce National Park' 'Yellowstone National Park'
```

```python
print(species_data['conservation_status'].unique())
#figure out most common types of species in each conservation status
print(species_data['category'].unique())
for con_status in species_data['conservation_status'].unique()[1:]:
    new_arr = []
    name_arr = []
    for cat in species_data['category'].unique():
        number = species_data[(species_data.category == cat) & (species_data.conservation_status == con_status)].size
        if number > 0:
            new_arr.append(number)
            name_arr.append(cat)
        #print(cat + '\t\t' + str(species_data[(species_data.category == cat) & (species_data.conservation_status == con_status)].size))
    plt.pie(new_arr, labels = name_arr)
    plt.title(con_status)
    plt.show()
```
